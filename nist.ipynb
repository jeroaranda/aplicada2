# Ajustar polinomio de grado 10

## Leer los datos y graficarlos

```python
nist=[]
with open('nist.txt') as f:
    for line in f:
        nist.append( [float(i) for i in line.split()])
nist=np.array(nist)
ax=sns.scatterplot(x=nist[:,1],y=nist[:,0],)
```

## Ajuste

```python
poly,cov=np.polyfit(x=nist[:,1],y=nist[:,0],deg=10,cov=True)
X=np.linspace(-9,-3,200)
y_pred=np.polyval(x=X,p=poly)
y_fit=np.polyval(x=nist[:,1],p=poly)
print(poly[::-1])
```
